```python
# GroupBy operation
data1 = {
    'Name' : ('Fahad', 'Anas', 'Yusuf'),
    'Age' : (20, 21, 22),
    'Address' : ('Nashik', 'Aligarh', 'Pratapgarh'),
    'Qualification' : ('BE', 'BE', 'BE')
    
}
```

```python
type(data1)
```

```python
# change this data into dataframe
df_new = pd.DataFrame(data1)
df_new
```

```python
# groupBy the data by name
df.groupby('Name')
print(df_new.groupby('Name').groups)
```

```python
gk = df_new.groupby('Name')
print(gk)
```

```python
# typecasting into tuple
tuple(gk)
```

```python
# iterating
for i in gk:
    print(i)
```

```python
# GOUPBY USING MULTIOPLE KEYS
# GROUPBY OPERATIONS
gg = df_new.groupby(['Name', 'Qualification'])
print(gg.groups)
```

```python
tuple(gg)
```

```python
for i in gg:
     print(i)
```
